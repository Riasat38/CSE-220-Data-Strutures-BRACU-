```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def isLeaf(node):
  if node.left == None and node.right == None:
    return True
  return False
```

```python
def convert_mirror(root):
  if isLeaf(root):
    return
  else:
    root.left , root.right = root.right,root.left
    convert_mirror(root.left)
    convert_mirror(root.right)
  return root


#DRIVER CODE
root = BTNode(10)
n1 = BTNode(20)
n2 = BTNode(30)
n3 = BTNode(40)
n4 = BTNode(60)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  40 20 60 10 30
print()

root2 = convert_mirror(root)
print('Mirrored Tree Inorder Traversal: ', end = ' ')
inorder(root2) #Mirrored Tree Inorder Traversal:  30 10 60 20 40
```

```python
def smallest_level(root,level={},lvl=0):
  if root == None:
    return level

  if lvl not in level:
    level[lvl] = root.elem
  else:
    if root.elem < level[lvl]:
      level[lvl] = root.elem
  smallest_level(root.left,level,lvl+1)
  smallest_level(root.right,level,lvl+1)
  return level


#DRIVER CODE
root = tree_construction([None, 4,9,2,3,-5,None,7])

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  3 9 5 4 2 7
print()
print('Level Wise Smallest Value: ', end = ' ')
print(smallest_level(root)) #Level Wise Smallest Value:  {0: 4, 1: 2, 2: -5}
```

```python
def LCA(root, x, y):

    if root == None:
        return None

    if root.elem > x and root.elem > y:
        return LCA(root.left, x, y)

    elif root.elem < x and root.elem < y:
        return LCA(root.right, x, y)

    else:
        return root.elem

#DRIVER CODE
#Write by yourself from the given tree
#check all the sample inputs given

root = BTNode(15)
n1 = BTNode(10)
n2 = BTNode(25)
n3 = BTNode(8)
n4 = BTNode(12)
n5 = BTNode(6)
n6 = BTNode(9)
n7 = BTNode(20)
n8 = BTNode(30)
n9 = BTNode(18)
n10 = BTNode(22)

n3.left , n3.right = n5, n6

n1.left , n1.right = n3 , n4

n7.left , n7.right = n9 , n10

n2.left , n2.right = n7 , n8

root.left ,root.right = n1 , n2

print(LCA(root,6,12))
print(LCA(root,20,6))
print(LCA(root,18,22))
print(LCA(root,20,25))
print(LCA(root,10,12))
```

```python
#better approach
def DCA(root,x,y):

  if root == None:
    return None

  left = DCA(root.left,x,y)
  right = DCA(root.right,x,y)

  if root.elem != x and root.elem != y:
    if left == None and right == None:
      return None
    if left != None and right == None:
      return root.left
    if right != None and left ==None:
      return root.right
    if left != None and right != None:
      return root

  else:
    return root


root = BTNode(15)
n1 = BTNode(10)
n2 = BTNode(25)
n3 = BTNode(8)
n4 = BTNode(12)
n5 = BTNode(6)
n6 = BTNode(9)
n7 = BTNode(20)
n8 = BTNode(30)
n9 = BTNode(18)
n10 = BTNode(22)

n3.left , n3.right = n5, n6

n1.left , n1.right = n3 , n4

n7.left , n7.right = n9 , n10

n2.left , n2.right = n7 , n8

root.left ,root.right = n1 , n2

print(DCA(root,6,12).elem)
print(DCA(root,20,6).elem)
print(DCA(root,18,22).elem)
print(DCA(root,20,25).elem)
print(DCA(root,10,12).elem)
```

```python
#WRITE THE FUNCTION and DRIVER CODE BY YOURSELF
def inOrderSum(node,sum=0):

  if node == None:
    return 0
  else:
    sum += inOrderSum(node.left,sum)
    sum += node.elem
    sum += inOrderSum(node.right,sum)
  return sum
def sumTree(root):
  if root == None:
    return 0
  left = inOrderSum(root.left)
  right = inOrderSum(root.right)

  if root.elem == left + right:
    return True
  if isLeaf(root):
    return True
  return False

root = BTNode(15)
n1 = BTNode(10)
n2 = BTNode(25)
n3 = BTNode(8)
n4 = BTNode(12)
n5 = BTNode(6)
n6 = BTNode(9)
n7 = BTNode(20)
n8 = BTNode(30)
n9 = BTNode(18)
n10 = BTNode(22)

n3.right , n3.left = n6 , n6

n1.left , n1.right = n3 , n4

n7.left , n7.right = n9 , n10

n2.left , n2.right = n7 , n8

root.left ,root.right = n1 , n2

print(sumTree(root))
```
