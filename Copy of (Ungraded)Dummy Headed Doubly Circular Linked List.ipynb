```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:

  # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
  def __init__(self):
    self.head = Node(None,None,None)
    self.head.next = self.head
    self.head.prev =self.head

  #create a DHDCL from an array
  def create_list(self, arr):

    for i in arr:
      temp = self.head
      while temp.next != self.head:
        temp = temp.next
      new = Node(i,self.head,temp)
      temp.next = new
      self.head.prev = new

    return self.head

  # Counts the number of Nodes in the list and return the number
  def countNode(self):

    count = 0
    temp = self.head.next
    while temp != self.head:
      temp=temp.next
      count +=1

    return count


  # prints the elementents in the list forward
  def forwardprint(self):

    temp = self.head.next
    while temp.next != self.head:
      print(temp.element,end = " --> ")
      temp=temp.next
    print(temp.element)

  # prints the elementents in the list backward
  def backwardprint(self):

    temp = self.head.prev
    while temp.next != self.head:
      print(temp.element,end = " <--")
      temp = temp.prev
    print(temp.element)


  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):

    if idx<0 or idx>(self.countNode()-1):
      return
    else:
      count = 0
      temp = self.head.next
      while temp !=self.head:
        if count == idx:
          return temp
        count += 1
        temp = temp.next


  # returns the index of the containing the given elementent. if the elementent does not exist in the List, return -1.
  def indexOf(self, element):

    idx = 0
    temp = self.head.next
    while temp != self.head:
      if temp.element == element :
        return idx
      temp = temp.next
      idx  += 1

    return -1


  # insert at specific index of the DHDCL
  def insert(self, element, idx):

    if idx == 0:
      temp = self.head.next
      new = Node(element,self.head.next,self.head)
      self.head.next = new
      temp.prev = new

    elif idx<0 or idx>self.countNode() :
      return
    else:
      before = self.nodeAt(idx-1)
      after = before.next
      new = Node(element,after,before)
      before.next = new
      after.prev=new

    return self.head


  #removes the node of the specific index and returns the elementent of the node
  def remove(self, idx):

    if idx == 0:
      removed = self.head.next
      after = removed.next
      self.head.next = after
      after.prev = self.head

    elif idx<0 or idx>(self.countNode()-1):
      return -1
    else:
      before = self.nodeAt(idx-1)
      removed = before.next
      before.next = removed.next
      removed.prev = before

    return removed.element


  def rotate_left(self,k):

    temp = self.head.next
    count = 1
    storage = DoublyList()
    i = 0
    while temp != self.head:
      if count <k:
        storage.insert(temp.element,i)
        i +=1

      if count == k:
        self.head.next = temp
        temp.prev = self.head
        break
      temp = temp.next
      count +=1
    #storage.forwardprint()

    t = self.head.next
    while t.next != self.head:
      t = t.next
    t.next = storage.head.next
    storage.head.prev.next = self.head
    self.head.prev = storage.head.prev

    return self.head


  def rotate_right(self, k):
    temp = self.head.next
    count = 1
    storage = DoublyList()
    i = 0
    while temp != self.head:
      if count <= k:
        storage.insert(temp.element,i)
        i +=1
      if count > k:
        self.head.next = temp
        temp.prev = self.head
        break
      temp = temp.next
      count +=1

    t = self.head.next
    while t.next != self.head:
      t = t.next
    t.next = storage.head.next
    storage.head.prev.next = self.head
    self.head.prev = storage.head.prev

    return self.head

  def swap(self,k):

    temp = self.head.next
    count = 1
    while temp != self.head:
      if count == k:
        c = 1
        t = temp.next
        while t != self.head:
          if c == count:
            t.element,temp.element = temp.element,t.element
          c+=1
          t=t.next

      count += 1
      temp = temp.next
    return self.head
```
