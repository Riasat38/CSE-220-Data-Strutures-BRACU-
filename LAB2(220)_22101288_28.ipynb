```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))
```

#Zigzag Walk

```python
def walk_zigzag(floor):
  row,col=floor.shape

  for y in range(col):
    if y%2==0:
      for x in range(0,row,2):
        print(floor[x][y],end=" ")
      print()
    if y%2!=0:
      if row%2==0:
        start=row-1
      if row%2!=0:
        start=row-2
      for x in range(start,-1,-2):
        print(floor[x][y],end= " ")
      print()




floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0'],
                  ['1' , '4' , '2' , '8' , '6']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print('################')
floor = np.array([[ '3' , '8' , '4' , '6' , '1'],
                  ['7' , '2' , '1' , '9' , '3'],
                  ['9' , '0' , '7' , '5' , '8'],
                  ['2' , '1' , '3' , '4' , '0']]
                )

print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District

```python
def create_fence(district, depth):
    district_row, district_col = district.shape
    new = np.zeros((district_row+depth + depth , district_col + depth),dtype=int)

    row,col = new.shape
    for x in range(row):
      for y in range(col):
        new[x][y]=8
      for k in range(district_row):
        for l in range(district_col):
          new[k+depth][l+depth]=district[k][l]
    return new

depth = 1
district = np.array([[2,3,4], [3,4,6], [2,1,4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
#This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print('################')
depth = 2
district = np.array([
                 [2,3,4,1],
                 [3,4,6,5],
                 [2,1,4,7]
                ])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats

```python
def strength_difference(clubroom):
  strength_diff = np.zeros(100, dtype = int)
  row,col = clubroom.shape
  idx=0
  for x in range((row//2)+1):
    for y in range(x,col):
      if x==y:
        continue
      else:
        strength_diff[idx]=clubroom[x][y]-clubroom[y][x]
        idx+=1
  return strength_diff[:idx]

clubroom = np.array([
[1,  2,  9,  7],
[4,  5,  1,  8],
[3,  6,  2,  7],
[2,  8,  6,  3]
])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print('Strength Difference Array is : ', returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle

```python
def atm_triangle(n):
  arr=np.zeros((n,n),dtype=int)

  for x in range(n):
    for y in range(x+1):
      if y==0  or y==x:
        arr[x][y]=x+1
      else:
        s=0
        for k in range(y,-1,-1):
          s+=arr[x-1][k]
        arr[x][y]=s
  return arr
def print_without_none(matrix):
  for i in range(len(matrix)):
    for j in range(i+1):
      print(matrix[i][j],end=' ')
    print()



n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
#This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

#Trace The BOT

```python
def axis(cmd,row,col):
  if cmd==1:
    col-=3
    row-=2
  if cmd==2:
    col-=1
    row-=2
  if cmd==3:
    col-=2
    row-=3
  if cmd==4:
    col-=2
    row-=1
  if cmd==5:
    col+=1
    row-=2
  if cmd==6:
    col+=3
    row-=2
  if cmd==7:
    col+=2
    row-=3
  if cmd==8:
    col+=2
    row-=-1
  if cmd==9:
    col-=3
    row+=2
  if cmd==10:
    col-=1
    row+=2
  if cmd==11:
    col-=2
    row+=1
  if cmd==12:
    col-=2
    row+=3
  return (row,col)

def moving_around(cmds):
  grid = np.full((7, 7), '.')
  grid[3][3]="-"
  x=3
  y=3
  for i in range(len(cmds)):
    row,col=axis(cmds[i],x,y)
    if (row>6 or col>6):
      continue
    elif (row<0 or col<0):
      continue
    else:
      x=row
      y=col
      grid[x][y]='*'
  grid[x][y]='/'


  return grid

cmds = np.array([5,11,2,9])
result = moving_around(cmds)
print_matrix(result)
#This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```
