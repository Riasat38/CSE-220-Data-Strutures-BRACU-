```python
class Patient:
    def __init__(self,id = None,name = None,age = None,bloodgroup = None,next = None,prev = None):
        self.id =id
        self.name = name
        self.age = age
        self.blood_group = bloodgroup
        self.next = next
        self.prev = prev


class WRM:
    def __init__(self):

       self.head = Patient()
       self.head.next = self.head
       self.head.prev = self.head

    def RegisterPatient(self,id,name,age,bloodgroup):

        patient = Patient(id,name,age,bloodgroup)
        if self.head.next!=self.head:
            temp = self.head
            while temp.next!=self.head:
                temp=temp.next
            temp.next = patient
            patient.prev = temp
            patient.next = self.head
            self.head.prev = patient
        else:
            self.head.next = patient
            self.head.prev = patient
            patient.next = self.head
            patient.prev = self.head

        print("Success!! Patient Registered.")
        return self.head

    def ServePatient(self):

        if self.head.next==self.head:
            print("No Patients to Serve.")
        else:
            served = self.head.next
            after = served.next
            self.head.next = after
            after.prev = self.head
            print(served.name,"is served.")

    def CancelAll(self):

        if self.head.next==self.head:
            print("No Patients to Cancel.")
        else:
            self.head.next = self.head
            self.head.prev = self.head
            print("Success!! All Appointments Canceled.")

    def CanDoctorGOHome(self):

        if self.head.next !=self.head:
            print("No!")
        else:
            print("Yes!")

    def ShowAllPatient(self):

        temp = self.head.next
        if temp.next == self.head:
            print("No Patients in Line.")
        else:
            while temp.next!=self.head:
                print(f"[<ID:{temp.id}, NAME:{temp.name}, AGE:{temp.age}, BLOOD GROUP:{temp.blood_group}>]",end=' <--> ')
                temp=temp.next
            print(f"[<ID:{temp.id}, NAME:{temp.name}, AGE:{temp.age}, BLOOD GROUP:{temp.blood_group}>]")

    def ReverseTheLine(self):

        if self.head.next == self.head:
            print("No Patients in Line.")

        current = self.head.next
        while current != self.head:
            temp = current.next
            current.next =current.prev
            current.prev =  temp
            current = current.prev
        temp =  self.head.next
        self.head.next =self.head.prev
        self.head.prev = temp

```

```python
L = WRM()
L.RegisterPatient(606,'Riasat',22,'O+')
L.RegisterPatient(707,'Zahin',18,'A+')
L.RegisterPatient(808,'Vuiya',25,'A-')
L.RegisterPatient(909,'Arko',13,'AB+')
L.ShowAllPatient()
L.ServePatient()
L.ServePatient()
L.CanDoctorGOHome()
L.ShowAllPatient()
L.CancelAll()
L.ShowAllPatient()
```
